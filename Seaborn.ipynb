```python
#categorical data
```

```python
sns.barplot(x='sex',y="total_bill",data=tips)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
#box plots and  violin plot
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,palette="rainbow")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,plette="rainbow")
```

```python
sns.violinplot(x="day",y="total_bill",hue='sex',data=tips,plette="rainbow")
```

```python
sns.violinplot(x="day",y="total_bill",hue='sex',data=tips,plette="rainbow",split=True)
```

```python
#strip plot and swarm plot
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue="sex",palette='Set1',dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue="sex",palette='Set1',dodge=True)
```

```python
#combining plot violin+swarm
```

```python
sns.violinplot(x='tip',y='day',palette="rainbow",data=tips)
sns.swarmplot(x='tip',y='day',data=tips,color='red')
```
