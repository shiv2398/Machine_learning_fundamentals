```python
from numpy.random import randn
import numpy as np
```

```python
df=pd.DataFrame(dict(time=np.arange(500),value=randn(500).cumsum()))
```

```python
df.head()
```

```python
sns.relplot(x='time',y='value',kind='line',data=df,sort=True)
```

```python
df=pd.DataFrame(randn(500,2).cumsum(axis=0),columns=['time','value'])
```

```python
sns.relplot(x='time',y='value',kind='line',data=df,sort=True)
```

```python
sns.relplot(x='time',y='value',kind='line',data=df,sort=False)
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,estimator=None)
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',hue="event",data=fmri)
```

```python
#with hue can change color
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,hue="region",style="event")
```

```python
#with hue can change color
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,hue="region",style="event",markers=True,dashes=False)
```

```python
#with hue can change color
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,hue="region",style="event")
```

```python
sns.relplot(x="timepoint",y='signal',hue="region",units="subject",estimator=None,kind="line",data=fmri)
```

```python
sns.relplot(x="timepoint",y='signal',hue="region",units="subject",estimator=None,kind="line",data=fmri.query("event=='stim'"))
```
