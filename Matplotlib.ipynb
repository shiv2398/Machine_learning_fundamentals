```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
output=x**2
```

```python
plt.plot(x,output,"b")
plt.xlabel("value of x")
plt.ylabel("value of y")
plt.title("square function")
```

```python
y=x**3
```

```python
plt.subplot(1,2,1)
plt.plot(x,output,"b--")
plt.title("square plot")
plt.subplot(1,2,2)
plt.plot(x,y,"k--")
plt.title("cubic ")
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0.2,0.7,0.7])
ax.plot(x,y,"b")
ax.set_title("cubic plot")
ax.set_xlabel(" x axis")
ax.set_ylabel("y label")

ax1=fig.add_axes([0.08,0.5,0.3,0.3])
ax1.plot(x,output,"y")
ax1.set_title("square root")
```

```python
#color ,Linewidth,linetype
```

```python
fig,ax=plt.subplots()
ax.plot(x,output,'g--')
ax.plot(x,y,'r*-')
```

```python
fig,ax=plt.subplots()
ax.plot(x,output,color="green")
ax.plot(x,y,'r*-')
```

```python
fig,ax=plt.subplots()
ax.plot(x,output,color="#FF0051")
ax.plot(x,y,'r*-')
```

```python
fig,ax=plt.subplots()
ax.plot(x,output,color="#FF0051",linewidth=10,linestyle="--")
ax.plot(x,y,'r*-',linewidth=5)
```

```python
fig,ax=plt.subplots()
ax.plot(x,output,color="#FF0051",linewidth=10,linestyle="--")
ax.plot(x,y,linewidth=5,marker='o',markersize=10,markerfacecolor="black",markeredgecolor="blue",linestyle="--")
```
