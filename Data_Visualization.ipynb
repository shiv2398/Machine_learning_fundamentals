```python
x=np.linspace(0,10,100)
fig=plt.figure()
plt.plot(x,np.sin(x),"-")
plt.plot(x,np.cos(x),"--")
```

```python
#subplots
```

```python
plt.figure()
plt.subplot(2,1,1)#(rows,columns,panel number)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,np.sin(x))
plt.subplot(1,2,2)
plt.plot(x,np.cos(x))
```

```python
plt.figure()
plt.subplot(3,3,1)
plt.plot(x,np.sin(x))

plt.subplot(3,3,2)
plt.plot(x,np.cos(x))

plt.subplot(3,3,3)
plt.plot(x,x**2)
plt.subplot(3,3,4)
plt.plot(x,np.tan(x))

plt.subplot(3,3,5)
plt.plot(x,x**3)
plt.subplot(3,3,6)
plt.plot(x,x**4)

plt.subplot(3,3,7)
plt.plot(x,x**5)
plt.subplot(3,3,8)
plt.plot(x,x**6)

plt.subplot(3,3,9)
plt.plot(x,x**7)
```

```python
#legends
```

```python
x=np.linspace(0,4*np.pi,1000)
plt.plot(x,x,label="linear",linestyle="--",linewidth=1,marker='*')
plt.plot(x,x**2,label="quadratic",color='c')
plt.xlabel("x_axis")
plt.ylabel("y axis")
plt.title("graph")
plt.legend()
```

```python
df=df.drop(["Date"],axis=1)
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.lineplot(x="Week_ID",y="Revenue",data=df)
```

```python
sns.lineplot(x="Week_ID",y="Revenue",data=df,hue="Promo")#hue used for categorical data and color based on the promo
```

```python

plt.figure(figsize=(20,10))
sns.lineplot(x="Week_ID",y="Revenue",data=df,hue="Promo",ci=None,markers=True,style="Promo")#hue used for categorical data and color based on the promo
```
