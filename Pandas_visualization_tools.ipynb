```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
from numpy.random import randn
```

```python
np.random.seed(0)
ts=pd.Series(randn(1000),index=pd.date_range('1/1/2019',periods=1000))
```

```python
ts1=ts
ts=ts.cumsum()
ts.plot()
```

```python
ts.head()
```

```python
 np.random.seed(0)
df=pd.DataFrame(randn(1000,4),index=ts.index,columns=['A','B','C',"D"])
```

```python
df=df.cumsum()
```

```python
df.head()
```

```python
df.plot(legend=False)
```
